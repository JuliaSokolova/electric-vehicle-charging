# Using bootstrapping to check out means distributions for tesla drivrs miles requests

```python
all_teslas_data = list(teslas.milesRequested)
len(all_teslas_data)
```

```python
def bootstrap_mean(data, n_samples = 10000):
    bootsrtap_means = []
    for i in range(n_samples):
        bootstrap_sample = np.random.choice(data, size = len(data), replace = True)
        bootsrtap_means.append(np.mean(bootstrap_sample))
    return bootsrtap_means
```

```python
all_tesla_boot_means = bootstrap_mean(all_teslas_data)
```

```python
ax, fig = plt.subplots(figsize = (18, 6))
ax = sns.distplot(all_tesla_boot_means)
#ax = plt.hist(all_tesla_boot_means)
```

```python
ax, fig = plt.subplots(figsize = (18, 6))
#ax = sns.distplot(all_tesla_boot_means)
ax = plt.hist(all_tesla_boot_means)
```

```python
#variance of bootstrapped means for all tesla users
all_teslas_var = np.var(all_teslas_data)
all_teslas_var_boot = np.var(all_tesla_boot_means)
boot_mean = np.mean(all_tesla_boot_means)

print('All tesla users sample mean: 43.532948')
print(f'All tesla users bootsrtap sample mean: {boot_mean}')

print(f'Variance of all tesla users charge request: {all_teslas_var}')
print(f'Variance of all tesla users means: {all_teslas_var_boot}')
```

```python
#confidence interval for all tesla users
left = np.percentile(all_tesla_boot_means, 2.5)
right = np.percentile(all_tesla_boot_means, 97.5)
left, right
```

# Using bootstrapping to check out means distributions for tesla drivrs kWh requests

```python
all_teslas_data_kWh = list(teslas.kWhRequested)
all_tesla_kWh_boot_means = bootstrap_mean(all_teslas_data_kWh)

ax, fig = plt.subplots(figsize = (18, 6))
ax = sns.distplot(all_tesla_kWh_boot_means)
```

```python
#variance of bootstrapped means for all tesla users
all_teslas_var_kWh = np.var(all_teslas_data_kWh)
all_teslas_var_boot_kWh = np.var(all_tesla_kWh_boot_means)
boot_mean_kWh = np.mean(all_tesla_kWh_boot_means)

print('All tesla users sample mean: 18.677977')
print(f'All tesla users bootsrtap sample mean: {boot_mean_kWh}')

print(f'Variance of all tesla users charge request: {all_teslas_var_kWh}')
print(f'Variance of all tesla users means: {all_teslas_var_boot_kWh}')
```
